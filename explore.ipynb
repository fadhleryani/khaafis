```python
import whisper_timestamped as whisper
import torch

model_size = "MohammedNasri/WHISPER-LARGE-ARABIC"
device = 'cuda' if torch.cuda.is_available() else 'cpu'
model = whisper.load_model(model_size,device=device)
audio = whisper.load_audio("sample_files/mas3ad_sample.wav")
segments = whisper.transcribe(model, audio, vad=True)
```

```python
from scipy.io import wavfile
from IPython.display import Audio

rate, signal = wavfile.read('sample_files/mas3ad_sample.wav')

def get_audio_segment(words):
    start = words[0]['start']
    end = words[-1]['end']
    return signal[int(round(start*rate, 0)):int(round(end*rate, 0))].T


# wordslice = slice(0,-1)
wordslice = slice(6,23) #pick random slice of words to test
words = segments['segments'][0]['words']
print([x['text'] for x in words[wordslice]])
Audio(get_audio_segment(words[wordslice]), rate=rate)
```
